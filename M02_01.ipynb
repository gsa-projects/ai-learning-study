```python
from math import *
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = torch.linspace(-pi, pi, 1000)
Y = torch.sin(X)
```

```python
X.shape
```

```python
a, b, c, d = torch.randn((4))

def model(x):
  return a * x**3 + b * x**2 + c*x + d
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

ax1.plot(X, Y)
ax1.set_title("Actual Y")

ax2.plot(X, model(X))
ax2.set_title("Predict Y")

plt.show()
```

$$
w := w - \eta \frac{\partial}{\partial w} \text{cost}(w)
$$

```python
lr = 1e-6
epochs = 2000

for epoch in range(epochs):
  predict = model(X)
  loss = torch.sqrt(torch.mean((predict - Y) ** 2))
  grad_loss = 2 * (predict - Y)
  
  a = a - lr * (grad_loss * X**3).sum()
  b = b - lr * (grad_loss * X**2).sum()
  c = c - lr * (grad_loss * X**1).sum()
  d = d - lr * (grad_loss * X**0).sum()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

ax1.plot(X, Y)
ax1.set_title("Actual Y")

ax2.plot(X, model(X))
ax2.set_title("Predict (Trained) Y")

plt.show()
```
