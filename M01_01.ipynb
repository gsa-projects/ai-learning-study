### 요약
사람이 학습한다는 말은 실생활에서 자주 쓰이지만, 기계가 학습한다는 말의 의미는 무엇인가? 기계의 학습을 의미하는 용어인 머신러닝은 문제를 수치화하여 기계에게 전달한 후, 인간이 학습하듯이 시행착오를 겪으며 수치화된 데이터에 가장 잘 맞는 함수를 찾아내는 것을 의미한다. 이러한 머신러닝의 방법에는 3가지가 있다. 데이터에 정답이 있는 '지도 학습', 데이터에 정답이 없는 '비지도 학습', 데이터 없이 인공지능이 직접 체험하며 학습하는 '강화 학습'이 그것이다. 지도 학습은 확실히 옳은 정답이 정해져 있는 문제에서 인풋이 옳은지 아닌지를 판단하는 데 사용한다. 비지도 학습은 데이터를 그룹으로 나누는 클러스터링에 사용된다. 강화 학습은 주어진 환경에 인공지능이 직접 상호작용해 스스로 개선되도록 할 때 사용된다.
