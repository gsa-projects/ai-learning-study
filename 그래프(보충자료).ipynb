```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
```

# 선 스타일

### **(1) 선 스타일 설정하기**

> 다음과 같이 line의 색상, 스타일, 굵기를 지정할 수 있습니다.
> * **line color** : `color` 인수에 이름을 넣어 색상 지정    


>* **line style** : `linestyle` 인수에 마크를 넣어 선스타일 지정


>* **line width** : `linewidth`인수에 숫자를 넣어 굵기를 지정

```python
## (1) 그래프의 배경 설정
plt.figure(figsize=(10, 6), facecolor='pink')  # 너비 10인치, 높이 6인치의 바탕색이 분홍인 figure를 생성
plt.title("꺾은선 그래프의 다양한 스타일", fontsize=25, fontweight='bold', pad=15)  # 제목을 생성
plt.xlabel("x축", fontsize=20)  # x축 이름을 생성
plt.ylabel("y축", fontsize=20, labelpad=30, rotation='horizontal')  # y축 이름을 설정하고 가로 방향으로 생성
plt.xlim(-0.5, 9.5)  # x축 범위를 -0.5부터 9.5로 설정
plt.ylim(-0.5, 9.5)  # y축 범위를 -0.5부터 9.5로 설정
plt.grid(axis='y', color='purple', linestyle=':')  # y축 눈금 격자를 보라색 점선으로 생성


## (2) 그래프의 스타일을 설정
plt.plot(range(10), range(10), color='b', linestyle='-', linewidth=2, label='blue solid')  # 파란색 solid 선그래프
plt.plot(range(6), [1, 2.5, 4, 5.5, 7, 8.5], color='g', linestyle='-.', linewidth=2, label='green dashdot')  # 초록색 dashdot 선그래프
plt.plot([2, 4, 6, 8], [0.5, 1.5, 2.5, 3.5], color='r', linestyle=':', linewidth=2, label='red dotted')  # 빨간색 dotted 선그래프
plt.plot(range(10), [9, 5, 3.5, 2.5, 2, 1.8, 1.5, 1.3, 1.2, 1.15], color='orange', linestyle='--', linewidth=3, label='orange dashed')  # 주황색 dashed 선그래프

plt.show()  # figure를 출력
```

### **(2) 선 위의 점 스타일 설정하기**

> 다음과 같이 marker의 모양과 크기를 지정할 수 있습니다.  
>
> * **marker** : `marker`인수에 마크를 넣어 모양 지정  

> * **marker size** : `markersize` 인수에 숫자를 넣어 크기를 지정 (선 위의 모든 데이터 marker가 같은 크기)

```python
## (1) 그래프의 배경 설정
plt.figure(figsize=(10, 6), facecolor='pink')  # 너비 10인치, 높이 6인치의 바탕색이 분홍인 figure를 생성
plt.title("꺾은선 그래프의 다양한 스타일", fontsize=25, fontweight='bold', pad=15)  # 제목을 생성
plt.xlabel("x축", fontsize=20)  # x축 이름을 생성
plt.ylabel("y축", fontsize=20, labelpad=30, rotation='horizontal')  # y축 이름을 설정하고 가로 방향으로 생성
plt.xlim(-0.5, 9.5)  # x축 범위를 -0.5부터 9.5로 설정
plt.ylim(-0.5, 9.5)  # y축 범위를 -0.5부터 9.5로 설정
plt.grid(axis='y', color='purple', linestyle=':')  # y축 눈금 격자를 보라색 점선으로 생성


## (2) 그래프의 스타일을 설정
plt.plot(range(10), range(10), color='b', linestyle='-', linewidth=2, marker='o', markersize=12, label='blue circle solid')  # 파란색 원 marker와 solid 선그래프
plt.plot(range(6), [1, 2.5, 4, 5.5, 7, 8.5], color='g', linestyle='-.', linewidth=2, marker='s', markersize=8, label='green square dashdot')  # 초록색 네모 marker와 dashdot 선그래프
plt.plot([2, 4, 6, 8], [0.5, 1.5, 2.5, 3.5], color='r', linestyle=':', linewidth=2, marker='D', markersize=8, label='red diamond dotted')  # 빨간색 다이아 marker와 dotted 선그래프
plt.plot(range(10), [9, 5, 3.5, 2.5, 2, 1.8, 1.5, 1.3, 1.2, 1.15], color='orange', linestyle='--', linewidth=3, marker='^', markersize=12, label='orange triangle_up dashed')  # 주황색 삼각형 marker와 dashed 선그래프

plt.show()  # figure를 출력
```

### **(3) 그래프의 범례 설정하기**
> 다음과 같이 `legend` 함수를 이용해 범례의 위치, 바탕색, 글씨색을 설정할 수 있습니다.  
> * **location** : `loc` 인수에 이름을 넣어 위치 지정  
|**upper left**|상단 왼쪽|**upper center**|상단 중앙|**upper right**|상단 오른쪽|
|**center left**|중앙 왼쪽|**center**|정중앙|**center right**|중앙 오른쪽|
|**lower left**|하단 왼쪽|**lower center**|하단 중앙|**lower right**|하단 오른쪽|
> * **facecolor** : `facecolor` 인수에 이름을 넣어 바탕색 지정  
> * **labelcolor** : `labelcolor` 인수에 이름을 넣어 글자색 지정  

```python
## (1) 그래프의 배경 설정
plt.figure(figsize=(10, 6), facecolor='pink')  # 너비 10인치, 높이 6인치의 바탕색이 분홍인 figure를 생성
plt.title("꺾은선 그래프의 다양한 스타일", fontsize=25, fontweight='bold', pad=15)  # 제목을 생성
plt.xlabel("x축", fontsize=20)  # x축 이름을 생성
plt.ylabel("y축", fontsize=20, labelpad=30, rotation='horizontal')  # y축 이름을 설정하고 가로 방향으로 생성
plt.xlim(-0.5, 9.5)  # x축 범위를 -0.5부터 9.5로 설정
plt.ylim(-0.5, 9.5)  # y축 범위를 -0.5부터 9.5로 설정
plt.grid(axis='y', color='purple', linestyle=':')  # y축 눈금 격자를 보라색 점선으로 생성


## (2) 그래프의 스타일을 설정
plt.plot(range(10), range(10), color='b', linestyle='-', linewidth=2, marker='o', markersize=12, label='blue circle solid')  # 파란색 원 marker와 solid 선그래프
plt.plot(range(6), [1, 2.5, 4, 5.5, 7, 8.5], color='g', linestyle='-.', linewidth=2, marker='s', markersize=8, label='green square dashdot')  # 초록색 네모 marker와 dashdot 선그래프
plt.plot([2, 4, 6, 8], [0.5, 1.5, 2.5, 3.5], color='r', linestyle=':', linewidth=2, marker='D', markersize=8, label='red diamond dotted')  # 빨간색 다이아 marker와 dotted 선그래프
plt.plot(range(10), [9, 5, 3.5, 2.5, 2, 1.8, 1.5, 1.3, 1.2, 1.15], color='orange', linestyle='--', linewidth=3, marker='^', markersize=12, label='orange triangle_up dashed')  # 주황색 삼각형 marker와 dashed 선그래프

## (3) 그래프의 범례를 설정
plt.legend(loc='upper right', facecolor='gray', labelcolor='w')  # 상단 오른쪽에 회색 바탕색의 하얀 글씨로 쓴 범례를 설정

plt.show()  # figure를 출력
```

# 산점도

### **(1) 모든 점의 특성 일괄 지정하기**

> * **marker** : `marker` 인수에 마크를 넣어 모양 지정  
> * **size** : `s` 인수에 숫자를 넣어 크기 지정
> * **color** : `c` 인수에 이름을 넣어 점 색상 지정
> * **edge color** : `edgecolors` 인수에 이름을 넣어 테두리 색상 지정

```python
## (1) 그래프의 배경 설정
plt.figure(figsize=(10, 6), facecolor='pink')  # 너비 10인치, 높이 6인치의 바탕색이 분홍인 figure를 생성
plt.title("산점도의 다양한 스타일", fontsize=25, fontweight='bold', pad=15)  # 제목을 생성
plt.xlabel("x축", fontsize=20)  # x축 이름을 생성
plt.ylabel("y축", fontsize=20, labelpad=30, rotation='horizontal')  # y축 이름을 설정하고 가로 방향으로 생성
plt.xlim(-0.5, 10)  # x축 범위를 -0.5부터 9.5로 설정
plt.ylim(-0.5, 9.5)  # y축 범위를 -0.5부터 9.5로 설정
plt.grid(axis='y', color='purple', linestyle=':')  # y축 눈금 격자를 보라색 점선으로 생성

## (2) 그래프를 그릴 데이터 생성
import numpy as np
X1 = 3 + np.random.standard_normal(100)  # 표준정규분포를 확률밀도함수로 하여 난수를 100개 생성하고 3을 더함 (평균=3)
Y1 = 1.5 * X1 + np.random.standard_normal(100)  # 표준정규분포를 확률밀도함수로 하여 난수를 100개 생성하여 Y1값을 계산

## (3) 그래프의 스타일 설정
plt.scatter(X1, Y1, c='yellow', s=200, marker="*", edgecolors='navy', label='yellow star navy edge')  # 노란색, 남색 테두리의 200 size 별모양 산점도를 그림

## (4) 그래프의 범례를 설정
plt.legend(loc='upper right', facecolor='gray', labelcolor='w')  # 상단 오른쪽에 회색 바탕색의 하얀 글씨로 쓴 범례를 설정

plt.show()  # figure를 출력
```

### **(2) 여러 점의 특성 각각 지정하기**
>
> * **size** : `s` 인수에 숫자 리스트를 넣어 숫자 크기에 따라 점 크기 지정
> * **color** : `c` 인수에 숫자 리스트를 넣어 숫자 크기에 따라 점 색상 지정

```python
## (1) 그래프의 배경 설정
plt.figure(figsize=(10, 6), facecolor='pink')  # 너비 10인치, 높이 6인치의 바탕색이 분홍인 figure를 생성
plt.title("산점도의 다양한 스타일", fontsize=25, fontweight='bold', pad=15)  # 제목을 생성
plt.xlabel("x축", fontsize=20)  # x축 이름을 생성
plt.ylabel("y축", fontsize=20, labelpad=30, rotation='horizontal')  # y축 이름을 설정하고 가로 방향으로 생성
plt.xlim(-0.5, 10)  # x축 범위를 -0.5부터 9.5로 설정
plt.ylim(-0.5, 9.5)  # y축 범위를 -0.5부터 9.5로 설정
plt.grid(axis='y', color='purple', linestyle=':')  # y축 눈금 격자를 보라색 점선으로 생성

## (2) 그래프를 그릴 데이터 생성
import numpy as np
X1 = 3 + np.random.standard_normal(100)  # 표준정규분포를 확률밀도함수로 하여 난수를 100개 생성하고 3을 더함 (평균=3, 표준편차=1)
Y1 = 1.5 * X1 + np.random.standard_normal(100)  # 표준정규분포를 확률밀도함수로 하여 난수를 100개 생성하여 Y1값을 계산

X2 = 5 + 2*np.random.standard_normal(50)  # 표준정규분포를 확률밀도함수로 하여 난수를 50 생성하고 2 곱한 후 5를 더함 (평균=5, 표준편차=2)
Y2 = 0.5*X2 + 0.5*np.random.standard_normal(50)  # 표준정규분포를 확률밀도함수로 하여 난수를 50개 생성하여 Y2값을 계산

## (3) 그래프의 스타일 설정
plt.scatter(X1, Y1, c='yellow', s=200, marker="*", edgecolors='navy', label='yellow star navy edge')  # 노란색, 남색 테두리의 200 size 별모양 산점도를 그림
plt.scatter(X2, Y2, s=X2**3, c=X2**2, label='size color by X2')  # 색상과 크기가 X2값에 따라 달라지는 산점도를 그림
plt.colorbar()  # 산점도의 색상이 나타내는 수치를 표현

## (4) 그래프의 범례를 설정
plt.legend(loc='upper right', facecolor='gray', labelcolor='w', markerscale=0.5)  # 상단 오른쪽에 회색 바탕색의 하얀 글씨로 쓴 범례를 설정, 범례에 들어갈 marker를 실제 크기의 0.5비율로 그림

plt.show()  # figure를 출력
```
