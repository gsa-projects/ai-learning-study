```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('penguins')
df.dropna(axis=0, inplace=True)
print(df.shape)
df.head()
```

```python
sns.pairplot(data=df, hue='sex')
```

```python
sns.histplot(data=df['species'])
```

```python
plt.figure(figsize=(16, 2))
plt.subplot(1, 3, 1)
sns.regplot(x='body_mass_g', y='bill_length_mm', data=df, scatter_kws={'alpha':0.2}, line_kws={'color': 'red'})
plt.subplot(1, 3, 2)
sns.regplot(x='body_mass_g', y='bill_depth_mm', data=df, scatter_kws={'alpha':0.2}, line_kws={'color': 'red'})
plt.subplot(1, 3, 3)
sns.regplot(x='body_mass_g', y='flipper_length_mm', data=df, scatter_kws={'alpha':0.2}, line_kws={'color': 'red'})
```

```python
sns.lmplot(x='body_mass_g', y='flipper_length_mm', data=df, hue='sex')
```

```python
sns.relplot(x='body_mass_g', y='bill_depth_mm', data=df, hue='sex')
```
